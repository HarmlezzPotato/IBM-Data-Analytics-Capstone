### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
null_values = df.isnull()
null_values
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
import numpy as np


plt.figure(figsize = (50,15))
sns.heatmap(data = null_values, cmap = "viridis")
ticks = np.arange(0,64260,5000)
plt.yticks(ticks = ticks, labels = ticks)
plt.yticks(fontsize = 15)
plt.xticks(fontsize = 15)
plt.xlabel("Columns", fontdict = {"fontsize":15})
plt.ylabel("Rows",fontdict = {"fontsize":15})
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
column_null_count = df["Employment"].isnull().sum()
column_null_count
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
frequent = df["Employment"].value_counts(ascending = False).reset_index()
most_frequent_value = frequent.iloc[0]
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
df["Employment"]= df.loc[:,"Employment"].fillna(most_frequent_value.iloc[0])
df["Employment"]
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
value = df["Employment"].value_counts().reset_index()[:21]
plt.figure(figsize = (20,10))
sns.barplot(data = value, x = "Employment", y = "count")
plt.xticks(rotation = 90)
plt.show
```
