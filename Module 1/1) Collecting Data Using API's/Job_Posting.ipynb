```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("job-postings.xlsx")

data = data.sort_values("Number Of Posts", ascending = False)

data.plot(kind = "bar", figsize = (15,9), x = "City", y = "Number Of Posts")
plt.title("Number of Job Postings In Each City")
plt.show()
```
