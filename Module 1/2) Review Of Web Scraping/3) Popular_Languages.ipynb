```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("popular-laguages.csv")

#The responses are not formatted properly, so I opted to split up the long responses from each respondent for easier analysis.
data = data.assign(Annaul_Average_Salary = data["Annaul_Average_Salary"].str.replace("$","").explode("Annaul_Average_Salary"))
data = data.assign(Annaul_Average_Salary = data["Annaul_Average_Salary"].str.replace(",","").explode("Annaul_Average_Salary"))

data["Annaul_Average_Salary"] = data["Annaul_Average_Salary"].astype("int")

data
```

```python
data = data.sort_values("Annaul_Average_Salary", ascending=False)

data.plot(kind = "bar", figsize = (15,9), x = "Language", y = "Annaul_Average_Salary")
plt.title("Average Annual Salary of Each Popular Programming Language")
plt.ylabel("Average Annual Salary ($)")
plt.show()
```
