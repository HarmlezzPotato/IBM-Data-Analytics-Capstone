### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
data = df[["DatabaseWantToWorkWith"]].dropna()

data = data.assign(DatabaseWantToWorkWith = data["DatabaseWantToWorkWith"].str.split(";").explode("DatabaseWantToWorkWith"))

data = data.value_counts(ascending= False)[:5].reset_index()

plt.figure(figsize = (10,9))
plt.pie(data = data, x = "count", labels= "DatabaseWantToWorkWith", autopct="%1.1f%%")
plt.title("Pie Chart of the Top 5 Databases Respondents Want to Work With")
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
data = df[["DevType"]].dropna()

data = data.value_counts(ascending = False).reset_index()[:5]

plt.figure(figsize = (10,9))
plt.pie(data = data, x = "count", labels= "DevType", autopct="%1.1f%%")
plt.title("Pie Chart of the Top 5 Developer Roles")
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here
data = df[["OpSysProfessional use"]].dropna()

data = data.assign(OpSysProfessional_use = data["OpSysProfessional use"].str.split(";").explode("OpsysProfessional_use"))

data = data.drop("OpSysProfessional use", axis = 1)

#I'll be doing top 10
data = data.value_counts(ascending = False).reset_index()[:10]

plt.figure(figsize = (15,14))
plt.pie(data = data, x = "count", labels= "OpSysProfessional_use", autopct="%1.1f%%")
plt.title("Pie Chart of the Top 10 Professional Operating System")
plt.show()
```
