```python
#!pip install pandas
#!pip install matplotlib

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

#### Step 1: Load the dataset


```python
#file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv("survey_data.csv")
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
## Write your code here
data = df[["Age","JobSatPoints_6"]].dropna()

np.random.seed(42)
def Transformation_Age(Age):
    if Age == "Under 18 years old":
        return np.random.randint(16,18)
    
    elif Age == "18-24 years old":
        return np.random.randint(18,25)
    
    elif Age == "25-34 years old":
        return np.random.randint(25,35)
    
    elif Age == "35-44 years old":
        return np.random.randint(35,45)
    
    elif Age == "45-54 years old":
        return np.random.randint(45,55)
    
    elif Age == "55-64 years old":
        return np.random.randint(55,65)
    
    elif Age == "65 years or older":
        return np.random.randint(65,81)
    
    else:
        return np.random.randint(16,81)
data["Age"] = data["Age"].map(Transformation_Age)

data.plot(kind = "scatter", figsize = (10,9), x = "Age", y = "JobSatPoints_6")
plt.title("Relstionship between Age and Job Satisfaction")
plt.show()
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
## Write your code here
data = df[["ConvertedCompYearly","JobSatPoints_6"]].dropna()

#I need to remove extreme outliers
data = data[data["ConvertedCompYearly"] < data["ConvertedCompYearly"].quantile(0.90)]

data.plot(kind = "scatter", figsize = (10,9), x = "JobSatPoints_6", y = "ConvertedCompYearly")
plt.title("Relationship Between Yearly Compensation and Job Satisfaction")
plt.show()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
## Write your code here
data = df[["Age","JobSatPoints_6"]].dropna()

np.random.seed(42)
def Transformation_Age(Age):
    if Age == "Under 18 years old":
        return np.random.randint(16,18)
    
    elif Age == "18-24 years old":
        return np.random.randint(18,25)
    
    elif Age == "25-34 years old":
        return np.random.randint(25,35)
    
    elif Age == "35-44 years old":
        return np.random.randint(35,45)
    
    elif Age == "45-54 years old":
        return np.random.randint(45,55)
    
    elif Age == "55-64 years old":
        return np.random.randint(55,65)
    
    elif Age == "65 years or older":
        return np.random.randint(65,81)
    
    else:
        return np.random.randint(16,81)
data["Age"] = data["Age"].map(Transformation_Age)

plt.figure(figsize = (10,9))
sns.regplot(data = data, x = "Age", y = "JobSatPoints_6", line_kws={"color":"red"})
plt.title("Relationship between Age and Job Satisfaction (With Regression line)")
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
## Write your code here
data = df[["Age","YearsCodePro"]].dropna()

np.random.seed(42)
def Transformation_Age(Age):
    if Age == "Under 18 years old":
        return np.random.randint(16,18)
    
    elif Age == "18-24 years old":
        return np.random.randint(18,25)
    
    elif Age == "25-34 years old":
        return np.random.randint(25,35)
    
    elif Age == "35-44 years old":
        return np.random.randint(35,45)
    
    elif Age == "45-54 years old":
        return np.random.randint(45,55)
    
    elif Age == "55-64 years old":
        return np.random.randint(55,65)
    
    elif Age == "65 years or older":
        return np.random.randint(65,81)
    
    else:
        return np.random.randint(16,81)
data["Age"] = data["Age"].map(Transformation_Age)

np.random.seed(43)
data["YearsCodePro"] = data["YearsCodePro"].replace({"Less than 1 year": str(0),
                            "More than 50 years": str(np.random.randint(51,80))})

data["YearsCodePro"] = data["YearsCodePro"].astype("int")


data.plot(kind = "scatter", figsize = (10,9), x = "Age", y = "YearsCodePro")
plt.title("Relationship between Age and Years of Coding Experience")
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
## Write your code here
data = df[["ConvertedCompYearly","JobSatPoints_6","Age"]].dropna()

#Removing extreme outliers
data = data[data["ConvertedCompYearly"] < data["ConvertedCompYearly"].quantile(0.99)]

np.random.seed(42)
def Transformation_Age(Age):
    if Age == "Under 18 years old":
        return np.random.randint(16,18)
    
    elif Age == "18-24 years old":
        return np.random.randint(18,25)
    
    elif Age == "25-34 years old":
        return np.random.randint(25,35)
    
    elif Age == "35-44 years old":
        return np.random.randint(35,45)
    
    elif Age == "45-54 years old":
        return np.random.randint(45,55)
    
    elif Age == "55-64 years old":
        return np.random.randint(55,65)
    
    elif Age == "65 years or older":
        return np.random.randint(65,81)
    
    else:
        return np.random.randint(16,81)
data["Age"] = data["Age"].map(Transformation_Age)

plt.figure(figsize = (15,14))
sns.scatterplot(data = data, x = "JobSatPoints_6", y = "ConvertedCompYearly", size = "Age", hue = "Age", palette="coolwarm", sizes = (20,300))
plt.title("Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size")
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here
data = df[["LanguageHaveWorkedWith","JobSatPoints_6"]].dropna()

#I need to split up the long responce in LanguageHaveWorkedWith
data = data.assign(LanguageHaveWorkedWith = data["LanguageHaveWorkedWith"].str.split(";").explode("LanguageHaveWorkedWith"))

plt.figure(figsize = (30,9))
sns.scatterplot(data = data, x = "LanguageHaveWorkedWith", y = "JobSatPoints_6", hue = "LanguageHaveWorkedWith", palette="tab10")
plt.xticks(rotation = 70)
plt.title("Scatter Plot for Popular Programming Languages by Job Satisfaction")
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here
data = df[["Employment","ConvertedCompYearly","JobSatPoints_6"]].dropna()

data = data.assign(Employment = data["Employment"].str.split(";").explode("Employment")).dropna()

data = data[data["ConvertedCompYearly"] < data["ConvertedCompYearly"].quantile(0.90)]

plt.figure(figsize = (20,13))
sns.scatterplot(data = data, x = "JobSatPoints_6", y = "ConvertedCompYearly", hue = "Employment", palette= "tab10")
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
## Write your code here
data = df[["Country","YearsCodePro","Age"]].dropna()

np.random.seed(42)
def Transformation_Age(Age):
    if Age == "Under 18 years old":
        return np.random.randint(16,18)
    
    elif Age == "18-24 years old":
        return np.random.randint(18,25)
    
    elif Age == "25-34 years old":
        return np.random.randint(25,35)
    
    elif Age == "35-44 years old":
        return np.random.randint(35,45)
    
    elif Age == "45-54 years old":
        return np.random.randint(45,55)
    
    elif Age == "55-64 years old":
        return np.random.randint(55,65)
    
    elif Age == "65 years or older":
        return np.random.randint(65,81)
    
    else:
        return np.random.randint(16,81)
data["Age"] = data["Age"].map(Transformation_Age)

np.random.seed(43)
data["YearsCodePro"] = data["YearsCodePro"].replace({"Less than 1 year": str(0),
                            "More than 50 years": str(np.random.randint(51,80))})

data["YearsCodePro"] = data["YearsCodePro"].astype("int")

plt.figure(figsize = (20,13))
sns.scatterplot(data = data, x = "Age", y = "YearsCodePro", hue = "Country", palette= "tab10")
plt.legend(bbox_to_anchor=(1,1), ncol=3)
plt.show()
```
